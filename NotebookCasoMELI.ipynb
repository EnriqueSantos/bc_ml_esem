```python
# Importamos las librerías necesarias
import requests
import pandas as pd
import json
import time
```

# Ejemplos del uso de la API

```python
# Información de categorías.

#Hacemos la solicitud a la API
cats = requests.get('https://api.mercadolibre.com/sites/MLM/categories')

# Convertir a DataFrame
df_cats = pd.DataFrame(cats.json())

df_cats.head(3)
```

```python
#Información de Items.

cat_id = 'MLA1000'
offset = 50
url  = f'https://api.mercadolibre.com/sites/MLA/search?category={cat_id}&offset={offset}'

items = requests.get(url)
items_json = items.json()

#Seleccionamos la información solo de los items desde la llave "results"

items_list_json = items_json['results']

#Convertimos en un dataframe
df_items = pd.DataFrame(items_list_json)

df_items.head(3)
```
