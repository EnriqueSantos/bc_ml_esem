# Definición de métricas

Para cada vendedor calculamos los percentiles 25 / 50 / 75 sobre el valor de sus productos. Esto nos permite evaluar en que rango de valor de valor se encuentra cada vendedor por transación.

```python
metrics_sellers = []
for s in df_items_cats['seller_id'].unique():
    seller_items = df_items_cats.loc[df_items_cats.seller_id==s]
    data_p25 = seller_items.price.quantile(0.25)
    data_p50 = seller_items.price.quantile(0.50)
    data_p75 = seller_items.price.quantile(0.75)
    try:
        total_transactions = df_reps.loc[df_reps['id']==s]['transactions'].values[0]
    except:
        total_transactions = ''
    metrics_sellers.append([s,data_p25,data_p50,data_p75,total_transactions])
```

```python
df_metrics = pd.DataFrame(metrics_sellers,columns=['seller','p25','p50','p75','total_transactions'])
```

Trabajamos solo con la muestra de datos de la que tenemos información completa:

```python
df_to_segment = df_metrics.loc[df_metrics.total_transactions!='']
```

Consideraremos los sellers con mayor p75, y mejor volumen de transactiones. Eso nos permite centrarnos primero en los sellers con mayor volumen en venta y por transacción.

```python
top_100_best_value_75 = df_to_segment.sort_values(['p75','total_transactions'],ascending=False)[0:100]
```
